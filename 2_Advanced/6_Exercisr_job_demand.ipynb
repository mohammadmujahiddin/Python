```python
import pandas as pd
from datasets import load_dataset
import matplotlib.pyplot as plt

# Loading data
dataset = load_dataset('lukebarousse/data_jobs')
df = dataset['train'].to_pandas()

# Data Cleaning
df['job_posted_date'] = pd.to_datetime(df['job_posted_date'])
```

```python
df_in = df[df['job_country'] == 'India'].copy()
```

```python
df_in
```

```python
df_in['job_posted_monthly'] = df_in['job_posted_date'].dt.strftime('%B')
df_in
```

```python
df_in_pivot = df_in.pivot_table(index='job_posted_monthly',columns='job_title_short', aggfunc='size')
```

```python
df_in_pivot.reset_index(inplace=True)
df_in_pivot['month_no'] = pd.to_datetime(df_in_pivot['job_posted_monthly'], format='%B').dt.month
df_in_pivot.sort_values('month_no', inplace=True)
df_in_pivot.set_index('job_posted_monthly',inplace=True)
df_in_pivot.drop(columns='month_no', inplace=True)
```

```python
df_in_pivot
```

```python
df_in_pivot.plot(kind='line')
```

```python
top_3 = df_in['job_title_short'].value_counts().head(3)
top_3 = top_3.index.tolist()
```

```python
df_in_pivot[top_3].plot(kind='line')
```
